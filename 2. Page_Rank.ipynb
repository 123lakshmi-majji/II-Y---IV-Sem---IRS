## 2. Implement how to analyze ranked retrieval of a very large number of documents with hyperlinks between them.

**Page Rank** Concept explained simply:

Each document is a node.

A hyperlink from Document A → Document B is a vote for B.

Pages with more votes (especially from important pages) get a higher rank.

This helps the IR system rank very large document sets.

```python
# Step 1: Represent documents and links (directed graph)
links = {
    "D1": ["D2", "D3"],     # D1 → D2, D3
    "D2": ["D3"],           # D2 → D3
    "D3": ["D1"],           # D3 → D1
    "D4": ["D3"],           # D4 → D3
}
```

```python
# Step 2: Initialize rank value for each document
N = len(links)
rank = {doc: 1 / N for doc in links}  # All docs start with equal rank
```

```python
# Parameters for PageRank
damping = 0.85     # Probability of following links
iterations = 10     # Repeat process for better accuracy
```

```python
# Step 3: PageRank computation
for i in range(iterations):
    new_rank = {}

    for doc in links:
        # Base rank if random jump happens
        new_rank[doc] = (1 - damping) / N

        # Add contributions from incoming links
        for other_doc in links:
            if doc in links[other_doc]:     # If other_doc → doc link exists
                new_rank[doc] += damping * (rank[other_doc] / len(links[other_doc]))

    rank = new_rank  # Update ranks
```

```python
# Step 4: Display final ranked results
print("Final Document Ranking Based on Hyperlinks:\n")
for doc, score in sorted(rank.items(), key=lambda x: x[1], reverse=True):
    print(f"{doc} : {score:.4f}")
```
